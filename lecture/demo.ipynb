## ndarray indexing

Basic accessing and modifying of ndarrays


```python
import numpy as np
```

```python
np.random.seed(0)
randNums = np.random.uniform(-5,5,size = (5,4))
```

```python
arr = randNums
```

```python
# arr[0, :]
arr[0]
```

```python
# access last 2 columns
print(arr)
print()
#print(arr[:, -2:]
#
```

```python
#access columns at indices 1-2 and in 1st row
# Low range (before :) contains that index
# High range (after :) does not contain that index (i-1)

print(arr)
print()
print(arr[0, 1:4])
```

```python
# Assign 1st row to -1s
arr[0, :] = -1
print(arr)
```

```python
# We want to reorder rows to swap 1st and last rows
arr = arr[[2,1,0],:]
print(arr)
```

```python
# arr[[0, -1], [0,2, 4]]
print(arr)
print(arr[np.ix_([0,-1], [0,2,4])])
```
