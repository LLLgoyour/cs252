## Switching around the axes of an ndarray and matrix multiplication in numpy


We can't matrix multiply the following ndarrays due to shape issues:


```python
a = 3*np.ones([3, 4])
b = 2*np.ones([3, 4])
# c = a @ b.T
c = a.T @ b
d = a @ b.T
print(a.T.shape, d.T.shape)
```

Need to pair up as

    (3, 4) x (4, 3)

OR

    (4, 3) x (3, 4)

Use the transpose to help out!


Note: Transposing a ndarray vector isn't meaningful if you don't have a singleton dimension


```python
a = np.ones(10)
a.shape
```

```python
a[np.newaxis, :].shape
```

```python
a[np.newaxis, :].T.shape
```

### Matrix vs. element-wise multiplication

- Star (\*) operator means element-wise multiplication
- Like other basic math operators, can use broadcasting (e.g. can multiply a (1,3) and a (5, 3) array)


```python
a1 = 2*np.ones([1, 3])
b1 = 3*np.ones([5, 3])
a1 @ b1.T
```

```python
print(np.sum(np.ones([1,2,3]),axis=1))
print(np.sum(np.ones([1,2,3]),axis=1).shape)
```

```python
print(np.mean(np.random.random([20,10]), axis =0))
print(np.mean(np.random.random([20,10]), axis =1).shape)
```

```python
print(np.zeros([5,5,5,5]))
print(np.zeros([5,5,5,5]).shape)
```

```python

print(np.squeeze(np.sum(np.zeros([5,4,2])[:,:,np.newaxis,:], axis = 0).shape))
```
