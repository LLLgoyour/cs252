## Task 2: Degree 3 Polynomial Regression

### 2a. Make 3 ndarrays

$$
\left(\begin{array}{cccc}
1 & x_0^1 & x_0^2 & x_0^3\\
1 & x_1^1 & x_1^2 & x_1^3\\
1 & x_2^1 & x_2^2 & x_2^3\\
1 & x_3^1 & x_3^2 & x_3^3\\
1 & x_4^1 & x_4^2 & x_4^3\\
... & ... & ... & ...\\
1 & x_{N-1}^1 & x_{N-1}^2 & x_{N-1}^3\\
\end{array}\right)
$$

1. Make a 2D polynomial ndarray for the fit dataset using the independent variable 2. Create a matrix of [ones](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ones.html), `A1f`, with the same number of rows as the fit dataset and 4 columns. 5. Update the second column to be the respective x column, the third column to be the respective x column squared, and the fourth column to be the respective x column cubed.
   1. Print the first 5 rows of the matrix
2. Make a 2D polynomial ndarray for the validation dataset 3. Create a matrix of [ones](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ones.html), `A1v`, with the same number of rows as the validation dataset and 4 columns 5. Like in the step above, update the second column to be the respective x column, the third column to be the respective x column squared, and the fourth column to be the respective x column cubed.
   1. Print the first 5 rows of the matrix
3. Make a 2D polynomial ndarray to graph the values on the regression curve.
   1. Create a 1D ndarray via [linspace](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html), `line_x`, spanning the range (i.e. min/max) of the validation dataset's x column values with 100 steps.
   2. Create a matrix of [ones](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ones.html), `x_p`, with the same number of rows as the length of `line_x` and 4 columns
   3. Update the second column to be `line_x`, the third column to be `line_x` squared, and the fourth column to be `line_x` cubed.
   4. Print the first 5 rows of the matrix


```python
### BEGIN SOLUTION
xf = Af[:, 0]
xv = Av[:, 0]

A1f = np.ones((Af.shape[0], 4))
A1f[:, 1] = xf
A1f[:, 2] = xf**2
A1f[:, 3] = xf**3
print('A1f first 5 rows')
print(A1f[:5])

A1v = np.ones((Av.shape[0], 4))
A1v[:, 1] = xv
A1v[:, 2] = xv**2
A1v[:, 3] = xv**3
print('A1v first 5 rows')
print(A1v[:5])

line_x = np.linspace(xv.min(), xv.max(), 100)
x_p = np.ones((line_x.shape[0], 4))
x_p[:, 1] = line_x
x_p[:, 2] = line_x**2
x_p[:, 3] = line_x**3
print('x_p first 5 rows')
print(x_p[:5])
### END SOLUTION

print('''
--------------------------------------
Your results should look like
--------------------------------------
A1f first 5 rows
[[ 1.       0.86084  0.74104  0.63792]
 [ 1.      -0.30754  0.09458 -0.02909]
 [ 1.       0.77201  0.596    0.46012]
 [ 1.      -0.71808  0.51564 -0.37027]
 [ 1.      -0.28735  0.08257 -0.02373]]
A1v first 5 rows
[[ 1.      -0.91878  0.84417 -0.77561]
 [ 1.      -0.92811  0.86139 -0.79946]
 [ 1.      -0.96837  0.93773 -0.90807]
 [ 1.      -0.96436  0.92999 -0.89685]
 [ 1.      -0.97014  0.94118 -0.91308]]
x_p first 5 rows
[[ 1.      -0.97014  0.94118 -0.91308]
 [ 1.      -0.94925  0.90107 -0.85534]
 [ 1.      -0.92835  0.86183 -0.80008]
 [ 1.      -0.90745  0.82347 -0.74726]
 [ 1.      -0.88656  0.78598 -0.69682]]
''')
```

### 2b. Calculate Degree 3 Polynomial Regression

1.  Use [scipy.linalg.lstsq]() to calculate and print the fitted coefficients and MSE for the dependent variable `y1` (i.e. dependent variable of the fit set).
2.  Print the $R^2$ value computed on the fit set.


```python
### BEGIN SOLUTION
c, _, _, _ = scipy.linalg.lstsq(A1f, y1f)
y1f_pred = A1f @ c
mse = np.mean((y1f - y1f_pred) ** 2)
r2 = 1 - np.sum((y1f - y1f_pred) ** 2) / np.sum((y1f - np.mean(y1f)) ** 2)

print('c:', c)
print('MSE:', mse)
print('r-squared:', r2)
### END SOLUTION

print('''
--------------------------------------
Your results should look like
--------------------------------------
c: [-0.15511 -0.96979 -0.24675  1.61541]
MSE: 0.3431335215559479
r-squared: 0.11595609001119211
''')
```

### 2c. Testing and prediction

1. Calculate the $R^2$ for the linear regressions ability predict the test dataset
   1. Multiply `A1v` (_validation set_) with the fitted coefficients to get the predicted values
   2. Calculate the $R^2$ value and print it computed with the _validation set_ the dependent variable values `yv`.
      > $R^2 = 1 - \frac{\sum_i \left (y_i - \hat{y}_i \right )^2}{\sum_i \left (y_i - \bar{y} \right )^2}$ <br> <br> $y_i$ are the dependent variable values from the validation dataset <br> $\bar{y}$ is the mean of the dependent variable values from the validation dataset <br> $\hat{y}_i$ is the y values _predicted_ by the regression

2. Print the $R^2$ for the test dataset


```python
### BEGIN SOLUTION
y1v_pred = A1v @ c
r2_validation = 1 - np.sum((y1v - y1v_pred) ** 2) / np.sum((y1v - np.mean(y1v)) ** 2)
print('r2 validation:', r2_validation)
### END SOLUTION

print('''
--------------------------------------
Your results should look like
--------------------------------------
r2 validation: 0.4431027176546438
''')
```

```python
### BEGIN SOLUTION
y_curve = x_p @ c

plt.figure(figsize=(10, 6))
plt.scatter(xf, y1f, label='Fit data', alpha=0.7)
plt.scatter(xv, y1v, label='Validation data', alpha=0.7, marker='x')
plt.plot(line_x, y_curve, label='Degree 3 regression curve', color='black', linewidth=2)

plt.title(f'Degree 3 Polynomial Regression (R² fit={r2:.4f}, R² val={r2_validation:.4f})')
plt.xlabel('x')
plt.ylabel('y1')
plt.legend()
### END SOLUTION
```
