```python
import numpy as np
# import matplotlib plotting library
import matplotlib.pyplot as plt
```

## Basic plotting example


### Plot log(x)


```python
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y = np.log(x)
# plot it!
plt.plot(x, y)
plt.show()
```

### Control the plot size

Specify the size with plt.figure() before creating the plot.


```python
plt.figure(figsize=(5,3))
x = np.linspace(1, 50)
# Evaluate the log
y = np.log(x)
# plot it!
plt.plot(x, y)
plt.show()
```

### Plot multiple log curves in same plot


```python
# We can plot a few more on the same plot
y_2 = np.log(20*x)
y_3 = np.log(30*x)

# plot them!
plt.plot(x, y)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.show()
```

### Customize colors


```python
# We can plot a few more on the same plot
y_2 = np.log(20*x)
y_3 = np.log(30*x)

# plot them!
plt.plot(x, y, 'b')
plt.plot(x, y_2, 'k')
plt.plot(x, y_3, 'g')
plt.show()
```

### Use a qualitative (categorical) color palette

Options here: https://matplotlib.org/stable/tutorials/colors/colormaps.html#qualitative

Access list of RGB colors via: `plt.cm.PALETTENAME.colors`


```python
plt.plot(x, y, c=plt.cm.tab10.colors[0])
plt.plot(x, y_2, c=plt.cm.tab10.colors[1])
plt.plot(x, y_3, c=plt.cm.tab10.colors[2])
plt.show()
```

### Customize examples: line thickness (aka width) and style


```python
plt.plot(x, y, c=plt.cm.tab10.colors[0], linewidth=3, linestyle='dashed')
plt.plot(x, y_2, c=plt.cm.tab10.colors[1],linewidth=5, linestyle='dotted')
plt.plot(x, y_3, c=plt.cm.tab10.colors[2],linewidth=7)
plt.show()
```

### Set title, axis viewing ranges, legend


```python
plt.plot(x, y, c=plt.cm.tab10.colors[0], linewidth=3, linestyle='dashed', label ='log1')
plt.plot(x, y_2, c=plt.cm.tab10.colors[1],linewidth=5, linestyle='dotted', label='log20')
plt.plot(x, y_3, c=plt.cm.tab10.colors[2],linewidth=7, label='log30')
plt.xlabel('My x axis')
plt.ylabel('My y axis')
plt.xlim([0, 60])
plt.ylim([0, 10])
plt.title('My pretty plot!')
plt.legend()
plt.show()
```
